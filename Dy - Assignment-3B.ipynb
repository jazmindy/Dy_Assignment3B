````python
## Code goes here
import pandas as pd
recweek = pd.read_csv("recweekdata2.csv")
profiles = pd.read_csv("studentprofiles.csv")
df = pd.merge(recweek,profiles, on = "lastname")
df = df.drop(columns=['firstname_y'])

# cellphone number 
df.loc[:,"phonenumber"] = df.loc[:,"phonenumber"].replace(r'^9',r'+63 9',regex=True)

# birthyear (include as type bc it wont read as a string unlike the phonenumber which has the +)
df.loc[:,"birthyear"] = df.loc[:,"birthyear"].astype(str).replace(r'^97',r'1997',regex=True)
df.loc[:,"birthyear"] = df.loc[:,"birthyear"].replace(r'^98',r'1998',regex=True)
df.loc[:,"birthyear"] = df.loc[:,"birthyear"].replace(r'^99',r'1999',regex=True)
df.loc[:,"birthyear"] = df.loc[:,"birthyear"].replace(r'^00',r'2000',regex=True)
df.loc[:,"birthyear"] = df.loc[:,"birthyear"].replace(r'^01',r'2001',regex=True)

# birthmonth
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].replace(r'(January|1|Jan)',r'January',regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].replace(r'(February|2|Feb|Febuaryruary)',r'February',regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].replace(r'(March|3|Mar)',r'March',regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].replace(r'(April|4|Apr)',r'April',regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].replace(r'(May|5|May)',r'May',regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].replace(r'(June|6|Jun)',r'June',regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].replace(r'(July|7|Jul)',r'July',regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].replace(r'(August|8|Aug)',r'August',regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].replace(r'(September|9|Sep)',r'September',regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].replace(r'(October|10|Oct|January0)',r'October',regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].replace(r'(November|11|JanuaryJanuary|Nov)',r'November',regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].replace(r'(December|12|JanuaryFebruary|Dec)',r'December',regex=True)

#birthday
df.loc[:,"birthday"]

#firstname
df.loc[:,"firstname_x"]

#lastname
df.loc[:,"lastname"]

# courses
df.loc[:,"course"] = df.loc[:,"course"].replace(r'(BS LM|Legal Management|LM)',r'Legal Management', regex=True)
df.loc[:,"course"] = df.loc[:,"course"].replace(r'(BS MGT|Management|MGT)',r'Management', regex=True)
df.loc[:,"course"] = df.loc[:,"course"].replace(r'(BS CTM|Communications Technology|CTM)',r'Communications Technology', regex=True)
df.loc[:,"course"] = df.loc[:,"course"].replace(r'(BS ITE|Information Technology Entrepreneurship|ITE)',r'Information Technology Entrepreneurship', regex=True)
df.loc[:,"course"] = df.loc[:,"course"].replace(r'(BS ME|Management Engineering|ME|Management Management Engineering)',r'Management Engineering', regex=True)
courses_title = df.loc[:,"course"]

# carrier
df.loc[:,"phonenumber"].str[0:7]
carrier = {"+63 917":"Globe","+63 918":"Smart", "+63 927":"Globe"}
df["carrier"] = df.loc[:,"phonenumber"].str[0:7].apply(lambda x: carrier[x])
carrier_title = df["carrier"]

df
````
